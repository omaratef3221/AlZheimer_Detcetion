```python
from main import *
```

```python
models = ["","VGG16", "efficientnet_v2", "ResNet50V2"]
```

```python
data = read_df()
images = read_images(data)
x_train, x_val, x_test, y_train, y_val, y_test = prepare_data(images, data)
```

## Models Training

```python
for each_model in models:
    gc.collect()
    gc.collect()
    if each_model == "":
        model = build_model()
        plot_model(model, to_file=('model_plot.png'), show_shapes=True, show_layer_names=True)
        train_evaluate_test(model, x_train, x_val, x_test, y_train, y_val, y_test, model_name = "Basic Model")
    else:
        model = build_model(model_name = each_model)
        plot_model(model, to_file= (each_model + '_plot.png'), show_shapes=True, show_layer_names=True)
        train_evaluate_test(model, x_train, x_val, x_test, y_train, y_val, y_test, model_name = each_model)
```
